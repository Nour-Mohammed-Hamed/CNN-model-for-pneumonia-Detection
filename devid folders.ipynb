```python
import pandas as pd 
import numpy as np 
import os
import shutil
import random
```

```python
file_covid_path='COVID-19_Radiography_Dataset/COVID.metadata.xlsx'

image_covid_path='COVID-19_Radiography_Dataset/COVID' #source where covid image store

image_normal_path='COVID-19_Radiography_Dataset/Normal' #source where covid image store


target_covid_path='COVID-19_Dataset/COVID' #Destination 

target_normal_path='COVID-19_Dataset/Normal' #Destination 
```

```python
df=pd.read_excel(file_covid_path)

df.head(4)
```

```python
covid_images=os.listdir(image_covid_path)#copy images and make it in list directory 

random.shuffle(covid_images) #rondom images 

# take 1500 images rondoml
for i in range(1,1501):
    
    covid_img=covid_images[i] #iterated in every image
    
    image_path=os.path.join(image_covid_path,covid_img) #intersection with the main folder 
    
    image_target=os.path.join(target_covid_path)
    
    shutil.copy2(image_path,image_target) #copy image from path to destination 
    
    print("copy image ",i) #print the copied image 
```

```python
normal_images=os.listdir(image_normal_path)#copy images and make it in list directory 

random.shuffle(normal_images) #rondom images 

# take 1500 images rondoml
for i in range(1,1501):
    
    notmal_img=normal_images[i] #iterated in every image
    
    image_path=os.path.join(image_normal_path,notmal_img) #intersection wiyh the main folder 
    
    image_target=os.path.join(target_normal_path)
    
    shutil.copy2(image_path,image_target) #copy image from path to destination 
    
    print("copy image ",i) #print the copied image 
```
